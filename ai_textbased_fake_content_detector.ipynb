```python
# Step 4: Feature extraction function
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import pos_tag
from nltk.probability import FreqDist

def extract_features(text):
    readability = textstat.flesch_reading_ease(text)
    sentences = sent_tokenize(text)
    lengths = [len(s.split()) for s in sentences]
    avg_length = np.mean(lengths) if lengths else 0
    std_dev = np.std(lengths) if lengths else 0

    tokens = word_tokenize(text)
    try:
        pos_tags = pos_tag(tokens)
    except LookupError:
        nltk.download('averaged_perceptron_tagger_eng')
        pos_tags = pos_tag(tokens)

    pos_counts = FreqDist(tag for (word, tag) in pos_tags)
    noun_ratio = pos_counts['NN'] / len(tokens) if 'NN' in pos_counts and len(tokens) > 0 else 0
    verb_ratio = pos_counts['VB'] / len(tokens) if 'VB' in pos_counts and len(tokens) > 0 else 0


    return [readability, avg_length, std_dev, noun_ratio, verb_ratio]
```

```python
# Step 5: Prepare data for training
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

X = []
y = []

if 'data_df' in locals() and not data_df.empty:
    for index, row in data_df.iterrows():
        features = extract_features(row["text"])
        X.append(features)
        y.append(1 if row["label"] == "AI" else 0)  # 1 = AI, 0 = Human
else:
    # Use original samples if CSV was not loaded
    for sample in samples:
        features = extract_features(sample["text"])
        X.append(features)
        y.append(1 if sample["label"] == "AI" else 0)  # 1 = AI, 0 = Human


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
# Step 6: Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# Step 7: Test model and print report
preds = model.predict(X_test)
print(classification_report(y_test, preds, target_names=["Human", "AI"]))
```

```python
# Step 8: Prediction function
def predict_text_origin(text):
    features = extract_features(text)
    prediction = model.predict([features])
    return "AI-Generated" if prediction[0] == 1 else "Human-Written"
```
